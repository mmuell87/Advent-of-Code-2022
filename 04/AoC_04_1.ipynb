```python
score=0
run=0
with open("input.txt") as f:
    while True:
        run += 1
#        print(run)
        line=[0,0,0,0]
        zeile=f.readline()
        if zeile=='':break
        zeile=zeile.replace(',','-',1)
        i=0
        for i in range(0,3,1):
            line[i]=int(zeile[0:zeile.index('-')])
            zeile=zeile.replace(str(line[i])+'-','',1)
        line[3]=int(zeile)
        if line[0]-line[2]<=0 and line[1]-line[3]>=0: score +=1
        elif line[0]-line[2]>=0 and line[1]-line[3]<=0: score += 1
#        if line[0]-line[2]==0 and line[1]-line[3]==0: score -=1
        print(str(run)+" "+str(line[0]-line[2])+" "+str(line[1]-line[3]) +" "+ str(score))
    print(str(score))
```
